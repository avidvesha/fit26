# Farmer Data


```python
import pandas as pd
import numpy as np
```

```python
locations = ['Garut', 'Indramayu', 'Karawang', 'Subang', 'Tasikmalaya']
```

```python
columns = ['dmu',	'Land area (m2)',	'Land lease value (IDR)',	'Labor cost (IDR)',	'Seed purchase value (IDR)',	'Fertilizer purchase value (IDR)',	'Pesticide purchase value (IDR)',	'Equipment rent value (IDR)',	'Production value (IDR)']
farmer = pd.read_excel('dataset.xlsx', engine='openpyxl', sheet_name='Farmer Data', header=None)
farmer = farmer.iloc[:, :9]
farmer.dropna(subset=[0], inplace=True)
farmer
```

```python
farmer = farmer[~farmer[0].isin(locations)]
farmer
```

```python
farmer.columns = columns
farmer
```

```python
farmer = farmer[farmer['dmu'] != 'dmu']
farmer
```

```python
locations_val = locations
location_map = []
location_idx = 0
first_row = True

for idx, row in farmer.iterrows():
    if int(row['dmu']) == 1 and not first_row:
        location_idx += 1
    
    first_row = False
    
    if location_idx < len(locations_val):
        location_map.append(locations_val[location_idx])
    else:
        location_map.append(None)

farmer['Location'] = location_map
farmer
```

```python
farmer['dmu'] = farmer['dmu'].astype(np.int64)

for col in farmer.columns[1:9]:
    farmer[col] = farmer[col].astype(np.float64)
farmer
```

```python
farmer.info()
```

```python
for loc in locations:
  print(loc)
  display(farmer[farmer.Location == loc].describe())
```
